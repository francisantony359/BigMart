```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory



dftrain = pd.read_csv("train_modified.csv")
dftest = pd.read_csv("test_modified.csv")
```

```python
combined = pd.concat([dftrain,dftest])
combined.head()
```

```python
combined["Item_Weight"].fillna(combined["Item_Weight"].mean(),inplace = True)
combined["Item_Outlet_Sales"].fillna(combined["Item_Outlet_Sales"].mean(),inplace = True)
print(combined["Item_Outlet_Sales"].isnull().sum())
print(combined["Item_Outlet_Sales"].isnull().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ("Item_Fat_Content", "Item_Identifier","Outlet_Identifier","Outlet_Location_Type","Outlet_Size", "Outlet_Type", "Item_Type"  )
for c in cols:
    label = LabelEncoder()
    label.fit(list(combined[c].values))
    combined[c] = label.transform(list(combined[c].values))
```

```python
combined.head()
```

```python
train_ID = dftrain["Item_Identifier"]
test_ID = dftest["Item_Identifier"]
```

```python
combined = combined.drop("Item_Identifier",axis =1)
```

```python
nrow_train = dftrain.shape[0]
X_train = combined[:nrow_train]
X_test = combined[nrow_train:]
```

```python
Y = dftrain["Item_MRP"]
combined = combined.drop("Item_MRP",axis = 1)
```

```python
combined.dtypes
```

```python
from sklearn.model_selection import KFold, cross_val_score, train_test_split
t_X, val_X, t_y, val_y = train_test_split(X_train,Y, test_size=0.25)
from xgboost import XGBRegressor

my_model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
my_model.fit(t_X,t_y)
predictions = my_model.predict(val_X)

from sklearn.metrics import mean_absolute_error
print("Mean Absolute Error : " + str(mean_absolute_error(val_y,predictions)))
```

```python
predictions = my_model.predict(X_test)
```

```python
my_submission = pd.DataFrame({'Id': test_ID, 'SalePrice': predictions})
# you could use any filename. We choose submission here
my_submission.to_csv('submission-3.csv', index=False)
```
